# Central Limit Theorem
## Graphing the distributions of means when rolling a d6

```python
import random
import matplotlib.pyplot as plt

# Roll a d6 num_rolls times
def rollX(num_rolls):
    rolls = {i:0 for i in range(1,7)}
    for _ in range(num_rolls):
        rolls[random.randint(1,6)] += 1
    return rolls

# Call rollX, n number of times and mean each result, returning all means in a list
def n_means(sample_size, num_samples=30000):
    result = []
    for _ in range(num_samples):
        rolls = rollX(sample_size)
        avg = 0
        for key, value in rolls.items():
            avg += key * value
        avg /= sample_size
        result.append(avg)
    return result
```

## Rolling 1 die 30,000 times
### Taking the average of a sample size of 1 is just returning that roll, so our resulting graph is identical to a frequency chart

```python
plt.hist(n_means(1), bins=6, align='left', rwidth=1);
```

## Rolling an increasing sample size of dice slowly conforms the graph to a normal distribution

```python
plt.hist(n_means(2), bins=20, align='left', rwidth=1);
```

```python
plt.hist(n_means(3), bins=20, align='left', rwidth=1);
```

```python
plt.hist(n_means(5), bins=26, align='left', rwidth=1);
```
